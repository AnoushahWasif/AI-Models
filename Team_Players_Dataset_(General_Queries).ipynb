What is the average number of passes made by forwards? By midfielders?
Make sure to label which is which.

```python
import numpy as np
```

```python
midfielder = []
for player in players:
  if player['position']=='midfielder':
    midfielder.append(int(player['passes']))
print("midfielder : ",np.average(midfielder))
```

```python
forward = []
for player in players:
  if player['position']=='forward':
    forward.append(int(player['passes']))
print("forward : ",np.average(forward))
```

Which team has the highest ratio of goalsFor to goalsAgainst?

Print the team only.

Reminder: Use float() to make sure you're doing floating point division

Hint: Use two variables to keep track of highest ratio seen so far
and team with that ratio:

**ratio = 0 # highest ratio seen so far**

**ratioteam = '' # team with highest ratio**

```python
ratio = 0
ratioteam = ''
for team in teams:
  if float(team['goalsFor'])/float(team['goalsAgainst']) > ratio:
    ratio = float(team['goalsFor'])/float(team['goalsAgainst'])
    ratioteam = team['team']
print(ratioteam, ratio)
```

How many players who play on a team with ranking <10 played more than 350 minutes?

Reminder: Convert ranking and minutes to integers before comparing to values

Hint: Compute join of players and teams, using a variable to count number of players satisfying requirement

```python
count = 0
for player in players:
  for team in teams:
    if player['team'] == team['team']:
      if int(team['ranking']) < 10 and int(player['minutes']) > 350:
        count += 1
print(count)
```
